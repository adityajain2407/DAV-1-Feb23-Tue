```python
'''
1-variable: univariate analysis
1. Numerical (N)
2. Categorical (C)

2-variable: bivariate analysis
N-N
C-C
N-C

2+ variable: Multivariate analysis
3 variable:
N N N
C C C
CCN
NNC
'''
```

```python
data.head()
```

```python
cat_counts = data["Genre"].value_counts()
```

```python
print(cat_counts)
```

```python
x_bar = cat_counts.index
print(x_bar)
```

```python
y_bar = cat_counts
```

```python
plt.bar(x_bar, y_bar)
```

```python
plt.figure(figsize = (12, 8))
plt.bar(x_bar, y_bar)
```

```python
plt.bar(x_bar, y_bar)
plt.xticks(rotation = 90, fontsize = 12)
```

```python
plt.bar(x_bar, y_bar, width = 0.2)
plt.xticks(rotation = 90, fontsize = 12)
```

```python
# plt.figure(figsize = (8, 5))
plt.bar(x_bar, y_bar, color = 'orange')
plt.title("Games per Genre", fontsize = 15)
plt.xlabel("Genre", fontsize = 12)
plt.ylabel("Count" , fontsize = 12)
plt.xticks(rotation = 90, fontsize = 15)
plt.yticks(fontsize = 12)
```

```python
plt.figure(figsize = (12, 5))
plt.bar(x_bar, y_bar, color = 'orange')
plt.title("Games per Genre", fontsize = 15)
plt.xlabel("Genre", fontsize = 12)
plt.ylabel("Count" , fontsize = 12)
plt.xticks(rotation = 90, fontsize = 15)
plt.yticks(fontsize = 12)
plt.show()
```

```python
sns.countplot(x = "Genre", data = data, color = 'orange')
plt.xticks(rotation = 90, fontsize = 12)
```

```python
sns.countplot(x = "Genre", data = data, order = data["Genre"].value_counts().index , color = 'orange')
plt.xticks(rotation = 90, fontsize = 12)
```
